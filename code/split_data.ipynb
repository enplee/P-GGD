```python
import os
from shutil import copy
import random
```

```python
def mkfile(file):
    if not os.path.exists(file):
        os.makedirs(file)
def split_data(datafile,dist_file,split_rate=0.1):
    classes = [cla for cla in os.listdir(datafile) if ".txt" not in cla]
    
    train_file = os.path.join(dist_file,"train")
    val_file = os.path.join(dist_file,"val")
    mkfile(train_file)
    for cla in classes:
        mkfile(os.path.join(train_file,cla))
    mkfile(val_file)
    for cla in classes:
        mkfile(os.path.join(val_file,cla))
    
    for cla in classes:
        cla_path = os.path.join(datafile,cla)
        images = os.listdir(cla_path)
        num = len(images)
        eval_index = random.sample(images,k = int(num*split_rate))
        for index,image in enumerate(images):
            if image in eval_index:
                image_path = os.path.join(cla_path,image)
                new_path = os.path.join(val_file,cla)
                copy(image_path,new_path)
            else:
                image_path = os.path.join(cla_path,image)
                new_path = os.path.join(train_file,cla)
                copy(image_path,new_path)
            print("\r[{}] processing [{}/{}]".format(cla, index+1, num), end="")
    print("")
```

```python
datafile = "./data/teaLevel/singleLeaf/"
dist_file = "./data/teaLevel/singleLeaf/"
split_data(datafile,dist_file,0.1)
```

```python
file = "./data/flower_data/flower_photos/" 
flower_class = [cla for cla in os.listdir(file) if ".txt" not in cla]
mkfile('flower_data/train')
for cla in flower_class:
    mkfile('./data/flower_data/train/'+cla)
mkfile('flower_data/val')
for cla in flower_class:
    mkfile('./data/flower_data/val/'+cla)
```

```python
split_rate = 0.1
for cla in flower_class:
    cla_path = file + cla + '/'
    images = os.listdir(cla_path)
    num = len(images)
    eval_index = random.sample(images, k=int(num*split_rate))
    for index, image in enumerate(images):
        if image in eval_index:
            image_path = cla_path + image
            new_path = './data/flower_data/val/' + cla
            copy(image_path, new_path)
        else:
            image_path = cla_path + image
            new_path = './data/flower_data/train/' + cla
            copy(image_path, new_path)
        print("\r[{}] processing [{}/{}]".format(cla, index+1, num), end="")  # processing bar
    print()

print("processing done!")
    
```
