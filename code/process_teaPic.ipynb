```python
import numpy as np
import cv2 as cv
import os
import cv2
```

```python
path = "./data/teaLevel/OneBud_Oneleaf_BXCtea/"
savePath = path+'res/'
images = os.listdir(path)
if not os.path.isdir(savePath):
    os.mkdir(savePath)
    
images_single = images[0:16]
images_screen = images[16:]

for imgPath in images_single:
    img = cv.imread(path+imgPath)
    shape = img.shape
    rowL = shape[1] // 6
    colL = shape[0] // 4
    cnt = 0
    for i in range(6):
        for j in range(4):
            cnt += 1
            roi = img[j*colL:(j+1)*colL,i*rowL:(i+1)*rowL]
            cv2.imwrite(+str(cnt)+"_from_"+imgPath,roi)
print("finish")
```

```python
def cut(path,savePath,start,end,row_n,col_n,batch=0):
    if not os.path.isdir(savePath):
        os.mkdir(savePath)
    images = os.listdir(path)
    print(len(images))
    for im in images:
        img = cv.imread(path+im)
        shape = img.shape
        rowL = shape[1] // row_n
        colL = shape[0] // col_n
        cnt = 0
        for i in range(row_n):
            for j in range(col_n):
                cnt+=1
                roi = img[j*colL:(j+1)*colL,i*rowL:(i+1)*rowL]
                cv2.imwrite(savePath+str(cnt)+"_"+str(batch)+"_from_"+im,roi)
        print("process: {}".format(im))
    print("finish")
```

```python
path = "E:/data/6/res/"
savePath = path+'res2/'
start = 0
end = 20
row_n = 3
col_n = 3

cut(path,savePath,start,end,row_n,col_n)
```

```python
def cut_margin(path,savePath,start,end,row1,row2,col1,col2):
    images = os.listdir(path)
    if not os.path.isdir(savePath):
        os.mkdir(savePath)
    for im in images:
        img = cv.imread(path+im)
        if len(img)==0: continue
        roi = img[col1:col2,row1:row2]
        cv2.imwrite(savePath+im,roi)
        print("process: {}".format(im))
    print("finish")
```

```python
def resize(path,savePath,size):
    images = os.listdir(path)
    if not os.path.isdir(savePath):
        os.mkdir(savePath)
    for im in images:
        img = cv.imread(path+im)
        roi = cv.resize(img,(size,size))
        cv2.imwrite(savePath+im,roi)
        print("process: {}".format(im))
    print("finish")
```

```python
path = "./data/teaLevel/singleLeaf/train/"
subDir = os.listdir(path)
for sd in subDir:
    p = path+sd+"/"
    resize(p,p,128)
    
```
